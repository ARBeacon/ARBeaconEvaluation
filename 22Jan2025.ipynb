```python
program_order = ["iBeacon-ARKit", "iBeacon-ARCore"]

import seaborn as sns
palette = sns.color_palette("deep", len(program_order))
```

```python
import pandas as pd

rows = []
for env_id, delays in ARKIT_SCAN_GROUP.getFirstObjectDelays().items():
    for delay in delays:
        rows.append({
            'environmentId': env_id,
            'delay': delay,
            'program': 'iBeacon-ARKit'
        })

for env_id, delays in ARCORE_SCAN_GROUP.getFirstObjectDelays().items():
    for delay in delays:
        rows.append({
            'environmentId': env_id,
            'delay': delay,
            'program': 'iBeacon-ARCore'
        })

delay_df = pd.DataFrame(rows)
delay_df.sort_values(by=['environmentId']).tail()
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(data=delay_df, x='environmentId', y='delay', hue='program', palette=palette, hue_order=program_order, alpha=1)

plt.title('Localization Delay')
plt.xlabel('Environment ID (Vision Features)')
plt.ylabel('Delay (sec)')
plt.legend(title='Approach')

plt.show()
```

```python
mean_df = delay_df.groupby(['environmentId', 'program'], as_index=False)['delay'].mean()
sns.scatterplot(data=mean_df, x='environmentId', y='delay', hue='program', palette=palette, hue_order=program_order, marker='X')

plt.title('Mean Localization Delay')
plt.xlabel('Environment ID (Vision Features)')
plt.ylabel('Delay (sec)')
plt.legend(title='Approach')

plt.show()
```

```python
rows = []
for env_id, distances in ARKIT_SCAN_GROUP.calculateDistance().items():
    for distance in distances:
        rows.append({
            'environmentId': env_id,
            'position_diff': distance['position'],
            'orientation_diff': distance['orientation'],
            'scale_diff': distance['scale'],
            'program': 'iBeacon-ARKit'
        })

for env_id, distances in ARCORE_SCAN_GROUP.calculateDistance().items():
    for distance in distances:
        rows.append({
            'environmentId': env_id,
            'position_diff': distance['position'],
            'orientation_diff': distance['orientation'],
            'scale_diff': distance['scale'],
            'program': 'iBeacon-ARCore'
        })

accuracy_df = pd.DataFrame(rows)
accuracy_df.sort_values(by=['environmentId']).tail()
```
