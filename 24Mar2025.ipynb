```python
avgObjectPoses = [scan.avgObjectPose() for scan in all_scans]
```

```python
def calculateRelativePose(avgObjectPose):
    imageRefs = avgObjectPose["imagesRefs"]
    bunny = avgObjectPose["bunny"]
    avgImageRef = ThreeDObject.avg(imageRefs.values())
    relative_original = bunny.relativePoseRefBy(avgImageRef)
    return relative_original

relativePoses = [calculateRelativePose(avgObjectPose) for avgObjectPose in avgObjectPoses]
relativePoses[:1]
```

```python
orientationDiffs = []
positionDiff = []

for relativePoseA in relativePoses:
    for relativePoseB in relativePoses:
        if relativePoseA == relativePoseB:
            continue
        diff = relativePoseA.distance(relativePoseB)
        orientationDiffs.append(diff["orientation"])
        positionDiff.append(diff["position"])

orientationDiffs = np.array(orientationDiffs)
positionDiff = np.array(positionDiff)

print(f"Maximum Position Diff:\t\t{max(positionDiff)}\tmeters")
print(f"Maximum Orientation Diff:\t{max(orientationDiffs)}\trad")

print(f"Mean Position Diff:\t\t{positionDiff.mean()}\tmeters")
print(f"Mean Orientation Diff:\t\t{orientationDiffs.mean()}\trad")
```
